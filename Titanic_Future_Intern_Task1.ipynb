# Imbalance

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind = "bar")
```

```python
X = df.drop('Survived',axis=1)
```

```python
y = df['Survived']
```

```python
from imblearn.over_sampling import SMOTE

# Instantiate SMOTE
smt = SMOTE(random_state=42)

# Fit and resample
X_smt, y_smt = smt.fit_resample(X, y)
```

```python
y_smt.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_smt,y_smt,test_size=0.20)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(rf_model,X_smt,y_smt,cv=20,n_jobs=-1)
```

```python
score
```

```python
score.mean()
```

```python
rfc_id3 = RandomForestClassifier(criterion='entropy')
```

```python
score1 = cross_val_score(rfc_id3,X_smt,y_smt,cv = 10)
```

```python
score1
```

```python
score1.mean()
```

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier


abc_model = AdaBoostClassifier()
rfc_model = RandomForestClassifier()
gbc_model = GradientBoostingClassifier()
dt_model = DecisionTreeClassifier()
lr_model = LogisticRegression()
kn_model = KNeighborsClassifier()
svc_model = SVC()
xg_model = XGBClassifier()

model = [abc_model, rfc_model, gbc_model, dt_model, lr_model,kn_model,svc_model, xg_model]

for m in model:
    cross_score =cross_val_score(m,X_smt,y_smt,cv = 10, n_jobs=-1)
    print(m)
    print(cross_score.mean())
    print()
    
```
